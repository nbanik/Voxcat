```python
import numpy as np
from matplotlib import pyplot as plt
from collections import defaultdict


#import the pixel edges

dec_bin_edges=np.loadtxt("/home/nil/LNmocks/Gauss_distribute/decbin_edges_20-60.txt")
ra_bin_edges=np.loadtxt("/home/nil/LNmocks/Gauss_distribute/rabin_edges_100.txt")
z_bin_edges=np.loadtxt("/home/nil/LNmocks/Gauss_distribute/zbin_edges.txt")  

for k in range(100):
    file = open("/home/nil/LNmocks/Gauss_distribute/vox_randoms/random_voxelized_cat{}.txt".format(k), "w")
    ran=np.loadtxt("/home/nil/LNmocks/Gaussphz/catalog{}.h5_random.dat".format(k))
    ran=ran[ran[:,0] >= 0.0]
    ran=ran[ran[:,0] <= 100.0]
    ran=ran[ran[:,1] >= 20.0]
    ran=ran[ran[:,1] <= 60.0]
    ran=ran[ran[:,2] >= 0.6]
    ran=ran[ran[:,2] <= 0.8]
    ran_ra=ran[:,0]
    ran_dec=ran[:,1]
    ran_z=ran[:,2]
    
    #Replace the ra and dec of the galxies with the mid value of ra and dec pixel in which they are located
    for i in range(len(ran_ra)):
        for j in range(len(ra_bin_edges)-1):
            if  ra_bin_edges[j] <= ran_ra[i] < ra_bin_edges[j+1]:
                ran_ra[i]=(ra_bin_edges[j] + ra_bin_edges[j+1])*0.5 
                  
               
    for i in range(len(ran_dec)):
        for j in range(len(dec_bin_edges)-1):
            if  dec_bin_edges[j]  <= ran_dec[i] < dec_bin_edges[j+1]:
                ran_dec[i]=(dec_bin_edges[j] + dec_bin_edges[j+1])*0.5 
               
    for i in range(len(ran_z)):
        for j in range(len(z_bin_edges)-1):
            if  z_bin_edges[j] <= ran_z[i] < z_bin_edges[j+1]:
                ran_z[i]=(z_bin_edges[j] + z_bin_edges[j+1])*0.5
    
    
    
    rdata=np.c_[ran_ra, ran_dec,ran_z]
    data_tuple=tuple(tuple(j) for j in rdata)
    dic = {}
    for sub in data_tuple:
        dic.setdefault(tuple(sub[:3]),[]).append(sub[:3])
    weight = defaultdict(int)
    for i in tuple(data_tuple):
        weight[i] += 1
    voxelized_data= np.c_[np.array(dic.keys()),weight.values()]

    print len(rdata), len(voxelized_data)
    ra=voxelized_data[:,0]
    dec=voxelized_data[:,1]
    z=voxelized_data[:,2]
    w=voxelized_data[:,3]
    print len(w)
    
    for index in range(len(z)):
        file.write(str(ra[index]) + "   " + str(dec[index])  + "   " + str(z[index]) + "   " + str(w[index]) +  "\n")
    file.close()
                              
    
    
    
```
